## 📌 Step 16: Feature and Target Split
- **X (features):** Sare columns except last (species)
- **y (target):** Only species column

````python
x=df.iloc[:,:-1]
y=df["spcies"]
````

## 📌 Step 17: Train-Test Split
- Data ko **training** aur **testing** parts me divide karte hain.
- Usually, 70–80% training aur 20–30% testing ke liye use hota hai.

````python
from sklearn.model_selection import train_test_split
````

## 📌 Step 17: Train-Test Split (with different random states)
- Data ko training aur testing me split karte hain.  
- Test size: **20%** data testing ke liye, 80% training ke liye.  
- Different `random_state` values try karne se data split alag hoga, lekin distribution same rahegi.  

````python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=17) #20 #13 #17 #14 #12
````

## 📌 Step 18: Import Random Forest Classifier
- Random Forest ek **ensemble learning algorithm** hai jo multiple decision trees banata hai.  
- Har tree data ke random subset par train hota hai aur final prediction **majority voting** se hoti hai.  
- Ye algorithm classification ke liye kaafi powerful hai.  

````python
from sklearn.ensemble import RandomForestClassifier
````

## 📌 Step 19: Initialize & Train Random Forest Classifier
- **n_estimators=500** → Forest mein total decision trees ki ginti (zyada trees → better accuracy).  
- **criterion="gini"** → Tree splitting ke liye Gini Index use hoga (impurity measure).  
- **max_depth=None** → Tree ki depth par koi restriction nahi.  
- **min_samples_split=2** → Node split karne ke liye minimum 2 samples required.  
- **min_samples_leaf=1** → Leaf node mein kam se kam 1 sample hona chahiye.  
- **max_features="sqrt"** → Har split par features ka random subset select hoga (sqrt of total features).  
- **bootstrap=True** → Har tree ko random sampling (with replacement) se train kiya jayega.  

Model ko train karne ke liye `.fit()` use karte hain.

````python
rfc = RandomForestClassifier( n_estimators=500, criterion="gini", max_depth=None, min_samples_split=2, min_samples_leaf=1, max_features="sqrt", bootstrap=True,)
rfc.fit(x_train,y_train)
````

## 📌 Step 20: Model Accuracy (Train vs Test)

- `rfc.score(x_test, y_test)` → Checks the accuracy on the **test dataset**.  
- `rfc.score(x_train, y_train)` → Checks the accuracy on the **training dataset**.  
- We multiply by `*100` to convert the values into percentages.  

✔️ **Test and Train values of this model:** `(97.0 , 100.0)`  

👉 This step helps us check whether the model is **overfitting** or not.  
- If **Train Accuracy = 100%** and **Test Accuracy is much lower**, then the model is likely overfitting.  
- Here, the gap is very small (97% vs 100%), which means the model might be **slightly overfitting**, but it still **generalizes well** to unseen data.  


````python
rfc.score(x_test,y_test)*100,rfc.score(x_train,y_train)*100
````

## 📌 Step 21: Adding Predictions to the Dataset

We can use our trained **Random Forest Classifier** to make predictions on the entire dataset `x` and store the results in a new column.  

````python
df["Prediction"]=rfc.predict(x)
df.head()
````

## 📌 Step XX: Cross-Validation (Model Stability Check)

- We applied **5-Fold Cross Validation** to check the stability and generalization of our Random Forest model.  
- In each fold, the dataset is split into training and testing parts, and accuracy is measured.  

✔️ **Cross Validation Scores (per fold):** `[0.9517, 0.9500, 0.9617, 0.9600, 0.9533]`  
✔️ **Mean Accuracy:** `95.53%`  
✔️ **Standard Deviation:** `0.46%`  

👉 Since the scores are very close to each other (low standard deviation), this shows that our model is **highly consistent, stable, and generalizes well** across unseen data.


````python
from sklearn.model_selection import cross_val_score
````

````python

cv_scores = cross_val_score(rfc, x, y, cv=5, scoring='accuracy')

print("Cross Validation Scores:", cv_scores)
print("Mean Accuracy:", cv_scores.mean()*100)
print("Standard Deviation:", cv_scores.std()*100)
````

## 📌 Step 23: Import Classification Metrics  

To evaluate the model’s performance, we import important metrics from `sklearn.metrics`:  

- **Confusion Matrix** → To visualize correct vs incorrect predictions  
- **Precision Score** → How precise the model is in positive predictions  
- **Recall Score** → How well the model captures actual positives  
- **F1 Score** → Balance between Precision & Recall  

````python
from sklearn.metrics import confusion_matrix,precision_score,recall_score, f1_score
````

## 📌 Step 24: Precision Score  

- **Precision** measures how many of the predicted positive cases are actually positive.  
- We use `average='weighted'` because our target variable has multiple classes (Iris species).  
- Multiplying by `100` gives the result in **percentage form**.  

````python
precision = precision_score(y_test, y_pred, average='weighted')*100
print("Precision Score:", precision)
````

## 📌 Step 25: Recall Score  

- **Recall** measures how many actual positive cases the model correctly identified.  
- We use `average='weighted'` to handle multiple classes fairly.  
- Multiplying by `100` gives the result in **percentage form**. 

````python
recall = recall_score(y_test, y_pred, average='weighted')*100
print("Recall Score:", recall)
````

## 📌 Step 26: F1 Score  

- **F1 Score** is the harmonic mean of **Precision** and **Recall**.  
- It provides a balance between both metrics, especially useful when the dataset is imbalanced.  
- We use `average='weighted'` for multi-class classification.  
- Multiplying by `100` gives the result in **percentage form**.

````python
f1 = f1_score(y_test, y_pred, average='weighted')*100
print("F1 Score:", f1)
````

## 📌 Step 27: Predictions on Test Data  

- Now we use the trained **Random Forest Classifier** to make predictions on the test dataset.  
- The `.predict(x_test)` function generates predicted labels for the unseen data.

````python
y_pred=rfc.predict(x_test)
y_pred
````

## 📌 Step 28: Confusion Matrix (Numerical Form)

- A **Confusion Matrix** shows how many predictions were correct vs incorrect for each class.  
- It is especially useful for evaluating classification models. 

````python
cm = confusion_matrix(y_test, y_pred)
cm
````

## 📌 Step 29: Confusion Matrix Heatmap  

- To better **visualize** the confusion matrix, we use a **heatmap**.  
- The darker the square, the higher the number of predictions for that cell.  
- X-axis → Predicted Labels  
- Y-axis → True Labels

````python
cm = confusion_matrix(y_test, y_pred)


plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix Heatmap")
plt.show()
````

## 📌 Step 30: Actual vs Predicted (Graphical Representation)

- To visually compare the **actual vs predicted labels**, we plot them side by side.  
- Each point represents a sample in the test dataset.  
- Black dots = **Actual Labels**  
- Blue crosses = **Predicted Labels**

````python
plt.figure(figsize=(10,6))
plt.scatter(range(len(y_test)), y_test, color="black", label="Actual")
plt.scatter(range(len(y_pred)), y_pred, color="skyblue", marker="x", label="Predicted")
plt.xlabel("Actual Species")
plt.ylabel("Species")
plt.title("Actual vs Predicted Species (Random Forest)")
plt.legend()
plt.grid(True)
plt.show()
````
